### Importing the dataset

For the prediction of credit_score (driver's risk score) using `linear regression`, 3 predictor variables will be used:

`1. Gender (gender) `<br>
`2. Age (agecat)` <br>
`3. Traffic Index (traffic_index)` <br>

We will extract each variable as a DataFrame

```python
# Import csv file as DF
data = pd.read_csv("Vehicle_policies_2020.csv")

# Extract the variables into DF
score = pd.DataFrame(data["credit_score"]) # Response 
gender = pd.DataFrame(data["gender"]) # Predictor
age = pd.DataFrame(data["agecat"]) # Predictor 
traffic = pd.DataFrame(data["traffic_index"])# Predictor

# Display data
print("Data dimensions: ", data.shape)
print("Data types:")
print(data.dtypes)
data.head()
```

# 2. Exploring Predictor Variables

```python
def plot(var):
    f, axes = plt.subplots(1, 3, figsize=(18, 9))
    sb.boxplot(data = var, orient = "h", ax = axes[0])
    sb.histplot(data = var, ax = axes[1])
    sb.violinplot(data = var, orient = "h", ax = axes[2])
```

## 2.1 Exploring gender

```python
sb.countplot(x='gender', data=data)
```

## 2.2 Exploring agecat


```python
plot(age)
```

## 2.3 Exploring traffic_index

```python
plot(traffic)
```

# 3. Exploring credit_score
